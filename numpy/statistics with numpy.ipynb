plt에 대해서

```python
'''
직선함수를 그려봅시다.
y = x + 15
'''
import matplotlib.pyplot as plt
import numpy as np
# 파이썬 코드
def func(x):
    return x+15

x = [i for i in range(100)]
y = []
for i in range(100):
    y.append(func(x[i]))
plt.plot(y,x)
```

```python
# numpy 코드

x = np.array([i for i in range(100)])
y = x+15
plt.plot(y,x)
```
